```python
import numpy as np
import pandas as pd

file_to_load = "Purchase_Data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

```python
#Total Player Count
player_count_df = purchase_data_df
player_count_df
```

```python
total_players = len(player_count_df["SN"].value_counts())
total_players

player_analysis = {'Total Players': [total_players]}
df_players = pd.DataFrame(player_analysis,index =[0])
df_players
```

```python
unique_value = player_count_df["Item Name"].nunique()
unique_value
average = player_count_df["Price"].mean()
average
total_purchase = player_count_df["Purchase ID"].count()
total_purchase
total_revenue = player_count_df["Price"].sum()
total_revenue
purchasing_analysis = {'Number of Unique Items': [unique_value],'Average Price': [average], 'Number of Purchases':[total_purchase],'Total Revenue':[total_revenue]}
df = pd.DataFrame(purchasing_analysis,index =[0])
df
df.style.format({'Average Price':"${:,.2f}",'Total Revenue': '${:,.2f}'})
```

```python
# GENDER ANALYSIS
```

```python
gender_df = player_count_df.drop_duplicates(subset='SN', keep='first')
gender_count = gender_df["Gender"].count()
gender_count
male_count = gender_df["Gender"].value_counts()['Male']
male_count
female_count = gender_df["Gender"].value_counts()['Female']
female_count
other_count = gender_count - male_count - female_count
other_count

male_percentage = male_count / gender_count * 100
female_percentage = female_count / gender_count * 100
other_percentage = other_count / gender_count * 100

gender_analysis = pd.DataFrame({"Genders" :['Female','Male','Other/Non-Disclosed'],
                   "Percentage of Players":[female_percentage, male_percentage, other_percentage],
                "Total Count":[female_count,male_count,other_count]})
gender_analysis

gender_analysis.style.format({'Percentage of Players':"{:,.2f}%"})
```
